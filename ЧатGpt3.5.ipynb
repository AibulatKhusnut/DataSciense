```python
import g4f 
response = g4f.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": "Сколько мне лет? как с помощью чата обрабатывать фотографии"}],
    stream=True,
)

for message in response:
    print(message, flush=True, end='')
```
